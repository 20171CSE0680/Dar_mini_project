Before spliting data for training and test dataset, there is need to replace outliers. There is several methods to deal with outliers, but we decided to replace observations that are below or above three standard deviation with the closest observation using function below.

```r
outliers <- function(x)
{
  d <- which(x<=(mean(x)-3*sd(x))) # down border
  u <- which(x>=(mean(x)+3*sd(x))) # up border
  if(length(d)>0) # replacing bottom observations if needed
  {
    m_d <- min(x[-d])
    x[d]<- m_d
  }
  if(length(u)>0) # replacing upper observations if needed
  {
    m_u <- max(x[-u])
    x[u]<-m_u
  }
  return(x) 
}
df2[,c(1,4,5,8,10)] <- apply(df2[,c(1,4,5,8,10)], 2, outliers)
```

First  method used is logistic regression. It is good to start from method that is more based on statistic and can be easily interpreted.

```r
model_log <- glm(data = df_train, target~., family = "binomial")#train
predictions <- factor(ifelse(predict(model_log, df_test, type = "response")>0.5,1,0))
summary(model_log)#model disc
print("training dataset")
confusionMatrix(df_train$target, factor(ifelse(model_log$fitted.values>0.5,1,0))) # confusion matrix for train data
print("test dataset")
confusionMatrix(df_test$target, predictions) # confusion matrix for test data
colAUC(predict(model_log, df_test, type = "response"), df_test$target, plotROC = T)# roc curve
```

Significantly important variables for building model were:

* sex
* chest pain
* thal
* number of major vessels

Accuracy for both datasets was about the same - around 80-88%. Specifity is a bit lower than accuracy, which suggest that model predictions are a bit unblanced and there is bigger chance of classyfing person with higher chance of heart attack as person with lower chance, than other way. Overall 80-88% accuracy is very good result.

Next method will be decision tree. This is another method that is easily interpretable. Results can be seen below:

image.png

```r
model1 <- rpart(data=df_train, target~.)
predictions1 <- predict(model1, df_test, type = "class")
print("training dataset")
confusionMatrix(df_train$target, predict(model1, df_train, type ="class"))
print("test dataset")
con1 <- confusionMatrix(df_test$target, predictions1)
con1
rpart.plot(model1, cex = 1.5)
colAUC(predict(model1, df_test, type = "prob"), df_test$target, plotROC = T)
```

Confusion matrix shows that less observations was predicted correctly in both test and training dasates than in logistic regression. This can also be seen on the ROC curve, where it is much further from the perfect line. Variables decisive for making decision were:

* chest pain
* thal
* number of major vessels
* slope

Big advantage of using decision tree is to see decision plot, where it is possible to check which variables and what value of them decided about division. On the other hand simple decision tree usually will perform worse than random forest method, which is similar method but instead of one it uses bigger number of trees, which is next method used.

```r
seeds <- as.vector(c(1:11), mode = "list")
seeds[[11]] <- 1

model_rf <- train(method ="rf", df_train[,-12], df_train$target,
                  metric = "Accuracy", trControl = trainControl(method = "repeatedcv", number =10, repeats = 10), seeds=seeds, tuneGrid = expand.grid(.mtry=c(2,3,4,5,8)))
model_rf
predictions2 <- predict(model_rf$finalModel, df_test, type = "class")
print("training dataset")
confusionMatrix(df_train$target, predict(model_rf$finalModel, df_train, type ="class"))
print("test dataset")
con2 <- confusionMatrix(df_test$target, predictions2)
con2
plot(model_rf)
colAUC(predict(model_rf$finalModel, df_test, type = "prob"), df_test$target, plotROC = T)
```

Model was trained by using repeated cross-valdiation, and tuning paramter mtry corresponding to number of variables used in each tree. Number of trees was set for default value. On the first plot it is possible to see what was average accuracy for each value of this parameter during run. The highest value was achieved for mtry equal 2 with about 82% of accuracy. Accuracy on the training set was 100% while for test set it was 86%, which suggest overfitting. Anyway 86% accuracy is very good score. Problem might be a bit low specifity - 83% this means 1 on 6 person with higher chance of heart attack wasn't detected.

Next method used for training model will be k nearest neighbours. This method is based on distance between osbervations, so it is important to scale observations, otherwise some variables with big values would have too much impact on the model. I decided to unitarize all numeric variables, using function showed in code below. Some categorical variables have more than 2 possible value, so they were converted to dummy variables, this process wasn't important for trees method, but for next methods it will be required.

```r
unitarization <- function(x)
{
  return(apply(x,2,function(y)
    {
    (y-min(y))/(max(y)-min(y))
  }))
}
df_train_dummy <- df_train[,c(2,3,7,9,10,11)]
df_train_dummy <- one_hot(as.data.table(df_train_dummy))
df_unit <- unitarization(df_train[,c(1,4,5,6,8)])
df_train_dummy <- cbind(df_train_dummy,df_unit)
df_train_dummy$target <- df_train$target

df_test_dummy <- df_test[,c(2,3,7,9,10,11)]
df_test_dummy <- one_hot(as.data.table(df_test_dummy))
df_unit2 <- unitarization(df_test[,c(1,4,5,6,8)])
df_test_dummy <- cbind(df_test_dummy,df_unit2)
df_test_dummy$target <- df_test$target
```

After preparing datasets, knn model can be trained:

```r
model_knn <- train(target ~ ., data = df_train_dummy, method = "knn", 
                  tuneGrid = expand.grid(.k = c(2,5,8,12,20,30)))
model_knn
predictions3 <- predict(model_knn$finalModel, df_test_dummy[,-25], type = "class")
print("training dataset")
confusionMatrix(df_train_dummy$target, predict(model_knn$finalModel, df_train_dummy[,-25], type ="class"))
print("test dataset")
con3 <- confusionMatrix(df_test_dummy$target, predictions3)
con3
plot(model_knn)
colAUC(predict(model_knn$finalModel, df_test_dummy[,-25], type = "prob"), df_test$target, plotROC = T)
```

In this method parameter to tune is k which corresponds to number of nearest neighbours predicted observation. Best performance was achieved for model with k equal 20. Accuracy on the test dataset was equal 83% which is good result but lower than random forest and logistic regression method. Similarly as in the previous models specificity is quite low comparing it to accuracy score.

Last method used for building model is support vector machine. Dataset used to build this model was same as for knn method. Prints and plots are shown below:

```r
set.seed(12)
model_svm <- train(method ="svmLinear", df_train_dummy[,-25], make.names(df_train$target),
                   metric = "Accuracy", trControl = trainControl(method = "repeatedcv", number =10, repeats = 10, classProbs =T
                                                            ),seeds=seeds,
                   tuneGrid = expand.grid(.C = seq(0, 5, length = 20)))
model_svm
print("training dataset")
levels(model_svm$finalModel@fitted) <- c("0","1")
confusionMatrix(df_train_dummy$target,model_svm$finalModel@fitted)
print("test dataset")
predictions4 <- predict(model_svm, df_test_dummy[,-25])
levels(predictions4) <- c("0","1")
con4 <- confusionMatrix(df_test_dummy$target, predictions4)
con4
plot(model_svm)
colAUC(predict(model_svm, df_test_dummy[,-25], type = "prob"), df_test$target, plotROC = T)
```
